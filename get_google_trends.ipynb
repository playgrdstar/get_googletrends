````python
import os
import time
from pathlib import Path
import pandas as pd
import numpy as np
from pytrends.request import TrendReq
````

````python
ticker_df = pd.read_csv('sp500_ticker_mappings.csv')
ticker_df['cashtag'] = '$' + ticker_df['TICKER']
ticker_df.head()
````

````python
pytrends = TrendReq(hl='en-US', tz=360)
````

````python
selected_ticker = [ticker_df.cashtag[2]]
print(selected_ticker)
````

````python
sel_ticker_trends = pytrends.get_historical_interest(selected_ticker, year_start=2020, month_start=1, day_start=1, hour_start=0, 
                                        year_end=2020, month_end=1, day_end=2, hour_end=0, 
                                        cat=0, geo='', gprop='', sleep=60)
````

````python
sel_ticker_trends
````

````python
kw_list = ['singapore', 'singaporean']
keyword_trends = pytrends.get_historical_interest(kw_list, year_start=2021, month_start=1, day_start=1, hour_start=0, 
                                        year_end=2021, month_end=1, day_end=2, hour_end=0, 
                                        cat=0, geo='', gprop='', sleep=60)
````
